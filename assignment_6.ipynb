#### 1) Verification and density

- **Support:** $x\in[0,1]$ (continuous on the whole real line with flat parts outside).
- **Limits:** $\lim_{x\to -\infty}F(x)=0$, $\lim_{x\to +\infty}F(x)=1$.
- **Monotonic & right‐continuous:** Nondecreasing; continuous at 0 and 1.

**PDF (derivative on $(0,1)$):**
$f(x)=\frac{d}{dx}\sqrt{x}=\frac{1}{2\sqrt{x}},\; 0<x<1,$  
and $f(x)=0$ otherwise. Note $f$ is integrable at $0$ and $\int_0^1 \frac{1}{2\sqrt{x}}\,dx=1$.


```python
import numpy as np
import matplotlib.pyplot as plt

def F1(x):
    x = np.asarray(x)
    out = np.zeros_like(x, dtype=float)
    mask_mid = (x>=0) & (x<=1)
    out[mask_mid] = np.sqrt(x[mask_mid])
    out[x>1] = 1.0
    return out

def f1(x):
    x = np.asarray(x)
    out = np.zeros_like(x, dtype=float)
    mask = (x>0) & (x<1)
    out[mask] = 1.0/(2.0*np.sqrt(x[mask]))
    return out

xs = np.linspace(-0.25, 1.25, 400)
xs_pdf = np.linspace(1e-6, 1-1e-6, 400) 

fig, axes = plt.subplots(1,2, figsize=(12,4))
axes[0].plot(xs, F1(xs), lw=2)
axes[0].set_title("CDF: F(x)")
axes[0].set_xlabel("x"); axes[0].set_ylabel("F(x)")
axes[0].grid(True, alpha=0.3)

axes[1].plot(xs_pdf, f1(xs_pdf), lw=2)
axes[1].set_title("PDF: f(x)")
axes[1].set_xlabel("x"); axes[1].set_ylabel("f(x)")
axes[1].set_ylim(bottom=0)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

#### 2) Verification and density

- **Support:** $\mathbb{R}$.
- **Limits:** $\lim_{x\to -\infty}F(x)=0$, $\lim_{x\to +\infty}F(x)=1$.
- **Monotonic & right‐continuous:** Smooth, strictly increasing.

**PDF:**
$f(x)=\frac{d}{dx}\frac{1}{1+e^{-x}}
=\frac{e^{-x}}{(1+e^{-x})^2}
=F(x)\bigl(1-F(x)\bigr),\; x\in\mathbb{R}.$


```python
def F2(x): return 1.0/(1.0 + np.exp(-x))
def f2(x): Fx = F2(x); return Fx*(1.0 - Fx)

xs = np.linspace(-8, 8, 600)
fig, axes = plt.subplots(1,2, figsize=(12,4))
axes[0].plot(xs, F2(xs), lw=2)
axes[0].set_title("CDF")
axes[0].set_xlabel("x"); axes[0].set_ylabel("F(x)")
axes[0].grid(True, alpha=0.3)

axes[1].plot(xs, f2(xs), lw=2)
axes[1].set_title("PDF")
axes[1].set_xlabel("x"); axes[1].set_ylabel("f(x)")
axes[1].grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
```

#### 3) Verification and density

- **Support:** $[a,b]$.
- **Limits:** $F(a^-)=0$, $F(b^+)=1$.
- **Monotonic & right‐continuous:** Continuous and increasing; differentiable except at $c$.

**PDF:**
$f(x)=\begin{cases}
\dfrac{2(x-a)}{(b-a)(c-a)}, & a<x<c,\\[6pt]
\dfrac{2(b-x)}{(b-a)(b-c)}, & c<x<b,\\[6pt]
0, & \text{otherwise.}
\end{cases}$  
$f$ is continuous at $x=c$ and integrates to 1.


```python
a, c, b = 0.0, 1.0, 2.0

def F3(x):
    x = np.asarray(x)
    out = np.zeros_like(x, dtype=float)
    mask1 = (x>=a) & (x<=c)
    mask2 = (x>=c) & (x<=b)
    out[mask1] = ((x[mask1]-a)**2)/((b-a)*(c-a))
    out[mask2] = 1.0 - ((b-x[mask2])**2)/((b-a)*(b-c))
    out[x>b] = 1.0
    return out

def f3(x):
    x = np.asarray(x)
    out = np.zeros_like(x, dtype=float)
    maskL = (x>a) & (x<c)
    maskR = (x>c) & (x<b)
    out[maskL] = 2.0*(x[maskL]-a)/((b-a)*(c-a))
    out[maskR] = 2.0*(b-x[maskR])/((b-a)*(b-c))
    return out

xs = np.linspace(a-0.5, b+0.5, 600)
fig, axes = plt.subplots(1,2, figsize=(12,4))
axes[0].plot(xs, F3(xs), lw=2)
axes[0].axvline(a, ls="--", alpha=0.4)
axes[0].axvline(c, ls="--", alpha=0.4)
axes[0].axvline(b, ls="--", alpha=0.4)
axes[0].set_title("CDF (Triangular)")
axes[1].plot(xs, f3(xs), lw=2)
axes[1].axvline(a, ls="--", alpha=0.4)
axes[1].axvline(c, ls="--", alpha=0.4)
axes[1].axvline(b, ls="--", alpha=0.4)
axes[1].set_title("PDF (Triangular)")
plt.tight_layout(); plt.show()
```

#### 4) Verification and density

- **Support:** $[0,\infty)$.
- **Limits:** $\lim_{x\to -\infty}F(x)=0$, $\lim_{x\to +\infty}F(x)=1$.
- **Monotonic & right‐continuous:** Standard exponential CDF.

**PDF:**
$f(x)=\frac{d}{dx}(1-e^{-x})=e^{-x},\; x>0;\; f(x)=0$ for $x<0.$  
(Define $f(0)=1$ as the right derivative; effectively $f(x)=e^{-x}\mathbf{1}\{x\ge0\}$.)


```python
def F4(x):
    x = np.asarray(x)
    out = np.zeros_like(x, dtype=float)
    mask = x>=0
    out[mask] = 1.0 - np.exp(-x[mask])
    return out

def f4(x):
    x = np.asarray(x)
    out = np.zeros_like(x, dtype=float)
    mask = x>=0
    out[mask] = np.exp(-x[mask])
    return out

xs = np.linspace(-2, 8, 600)
fig, axes = plt.subplots(1,2, figsize=(12,4))
axes[0].plot(xs, F4(xs), lw=2)
axes[0].set_title("CDF")
axes[1].plot(xs, f4(xs), lw=2)
axes[1].set_title("PDF")
plt.tight_layout(); plt.show()
```

$$
f(k;p_1,...,p_K) = p_1^{k=1}p_2^{k=2}...p_K^{k=K}
$$

## Logistic Distribution

The support of a logistic distribution is the set of all real numbers:

Support: x∈[−∞,∞]


**Probability Density Function (PDF):**

Differentiate with respect to \(x\):

$
\begin{aligned}
f(x; \sigma)
&= \frac{d}{dx} \left( \frac{1}{1 + e^{-x/\sigma}} \right) \\[6pt]
&= \frac{e^{-x/\sigma}}{\sigma \left( 1 + e^{-x/\sigma} \right)^2}
\end{aligned}
$

$
\boxed{
f(x; \sigma) = 
\frac{e^{-x/\sigma}}{\sigma (1 + e^{-x/\sigma})^2}, 
\quad x \in (-\infty, \infty), \; \sigma > 0
}
$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp

# Parameters
sigmas = [1, 2, 3, 4, 5]
grid = np.linspace(-10, 10, 400)

colors = sns.color_palette("coolwarm", len(sigmas))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Loop over σ values
for s, c in zip(sigmas, colors):
    pdf = sp.stats.logistic.pdf(grid, loc=0, scale=s)
    cdf = sp.stats.logistic.cdf(grid, loc=0, scale=s)
    
    sns.lineplot(x=grid, y=pdf, ax=axes[0], label=f"σ={s}", color=c)
    sns.lineplot(x=grid, y=cdf, ax=axes[1], label=f"σ={s}", color=c)

# Titles and labels
axes[0].set_title("Logistic PDF (Density)")
axes[1].set_title("Logistic CDF (Distribution)")
axes[0].set_xlabel("x")
axes[1].set_xlabel("x")
axes[0].set_ylabel("Density")
axes[1].set_ylabel("Cumulative Probability")

axes[0].legend(title="σ (scale)")
axes[1].legend(title="σ (scale)")

plt.tight_layout()
plt.show()
```

## Exponential Distribution

The support of an exponential distribution is the set of all nonnegative real numbers:

Support: $x \in [0, \infty]$

**Probability Density Function (PDF):**

Differentiate with respect to \(x\):

$
\begin{aligned}
f(x; \lambda)
&= \frac{d}{dx} \left( 1 - e^{-\lambda x} \right) \\[6pt]
&= \lambda e^{-\lambda x}
\end{aligned}
$

$
\boxed{
f(x; \lambda) = 
\lambda e^{-\lambda x}, 
\quad x \in [0, \infty], \; \lambda > 0
}
$


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp

# Parameters
lambdas = [0.5, 1, 1.5, 2, 2.5]
mu = 0
grid = np.linspace(0, 5, 400) 

colors = sns.color_palette("coolwarm", len(lambdas))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Loop over λ values
for lam, c in zip(lambdas, colors):
    pdf = sp.stats.expon.pdf(grid, loc=mu, scale=1/lam)   # scale = 1/λ
    cdf = sp.stats.expon.cdf(grid, loc=mu, scale=1/lam)
    
    # Plot both on the same axes
    sns.lineplot(x=grid, y=pdf, ax=axes[0], label=f"λ={lam}", color=c)
    sns.lineplot(x=grid, y=cdf, ax=axes[1], label=f"λ={lam}", color=c)

# Titles and labels
axes[0].set_title("Exponential PDF (Density)")
axes[1].set_title("Exponential CDF (Distribution)")
axes[0].set_xlabel("x")
axes[1].set_xlabel("x")
axes[0].set_ylabel("Density")
axes[1].set_ylabel("Cumulative Probability")

axes[0].legend(title="λ (rate)")
axes[1].legend(title="λ (rate)")

plt.tight_layout()
plt.show()
```

## Negative Binomial

The support of a negative binomial distribution is the set of all nonnegative integers:

Support: $x \in \{0, 1, 2, 3, \dots\}$


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp

# Parameters
r_values = [2, 4, 6, 8]         
p_values = [0.2, 0.4, 0.6, 0.8] 
colors = sns.color_palette("coolwarm", len(r_values))

x = np.arange(0, 40)      # support: {0, 1, 2, ...}

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Loop over parameter pairs and colors
for r, p, c in zip(r_values, p_values, colors):
    pmf = sp.stats.nbinom.pmf(x, r, p)
    cdf = sp.stats.nbinom.cdf(x, r, p)

    # PMF: use stem for discreteness
    markerline, stemlines, baseline = axes[0].stem(x, pmf, basefmt=' ')
    plt.setp(markerline, 'color', c)
    plt.setp(stemlines, 'color', c)
    axes[0].plot(x, pmf, color=c, label=f"r={r}, p={p}")

    axes[1].step(x, cdf, where='post', color=c, label=f"r={r}, p={p}")

axes[0].set_title("Negative Binomial PMF (failures before r successes)")
axes[1].set_title("Negative Binomial CDF")
axes[0].set_xlabel("k (failures)")
axes[1].set_xlabel("k (failures)")
axes[0].set_ylabel("Probability")
axes[1].set_ylabel("Cumulative Probability")
axes[0].legend()
axes[1].legend()
plt.tight_layout()
plt.show()
```

## Categorical

Categorical (similar to Bernoulli): The probability mass function over $k = 1, 2, ..., K$ categories is
$$
f(k;p_1,...,p_K) = p_1^{k=1}p_2^{k=2}...p_K^{k=K}
$$
where $0 \le p_i \le 1$ and $ \sum_{k=1}^K p_k = 1$.

The support of a negative binomial distribution is the set of all possible discrete category labels.


```python
categories = np.arange(1, 6)  # k = 1,...,K
params = [
    [1/5]*5,                         # uniform
    [0.50, 0.20, 0.15, 0.10, 0.05], # skewed right
    [0.05, 0.05, 0.10, 0.30, 0.50], # skewed left
]
colors = sns.color_palette("coolwarm", len(params))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for p, c in zip(params, colors):
    p = np.array(p)
    F = np.cumsum(p)
    axes[0].stem(categories, p, basefmt=' ', linefmt=c[0:3], markerfmt='o', label=f"p={np.round(p,2)}")
    axes[1].step(categories, F, where='post', color=c, label=f"p={np.round(p,2)}")

axes[0].set_title("Categorical PMF (Theoretical)")
axes[1].set_title("Categorical CDF (Theoretical)")
axes[0].set_xlabel("Category k")
axes[1].set_xlabel("Category k")
axes[0].set_ylabel("Probability")
axes[1].set_ylabel("Cumulative Probability")
axes[0].legend(fontsize=8)
axes[1].legend(fontsize=8)
plt.tight_layout()
plt.show()
```

```python
p = np.array([0.50, 0.20, 0.15, 0.10, 0.05])  # skewed right
n = 1000
rng = np.random.default_rng(123)
sample = rng.choice(categories, size=n, p=p)

# Empirical PMF and CDF
counts = np.bincount(sample, minlength=len(categories)+1)[1:]
pmf_emp = counts / counts.sum()
cdf_emp = np.cumsum(pmf_emp)

# Theoretical PMF and CDF
pmf_th = p
cdf_th = np.cumsum(p)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# PMF
axes[0].bar(categories, pmf_emp, alpha=0.4, label="Empirical PMF (n=1000)")
axes[0].plot(categories, pmf_th, marker='o', color='C1', label="Theoretical PMF")
axes[0].set_title("Categorical PMF: Empirical vs Theoretical (Skewed Right)")
axes[0].set_xlabel("Category k")
axes[0].set_ylabel("Probability")
axes[0].legend()

# CDF
axes[1].step(categories, cdf_emp, where='post', label="Empirical CDF")
axes[1].step(categories, cdf_th, where='post', color='C1', label="Theoretical CDF")
axes[1].set_title("Categorical CDF: Empirical vs Theoretical (Skewed Right)")
axes[1].set_xlabel("Category k")
axes[1].set_ylabel("Cumulative Probability")
axes[1].legend()

plt.tight_layout()
plt.show()
```
